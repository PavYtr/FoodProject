```python
import os
import cv2
import numpy as np
import yaml
import shutil
from tqdm import tqdm
from pathlib import Path
from ultralytics import YOLO
```

```python
model = YOLO('yolo26s-cls.pt')
model_size = os.path.getsize('yolo26s-cls.pt') / 1e6
print(f"Размер модели: {model_size:.1f} MB")
```

```python


results = model.train(
    data='C:/training/yolo_food101_ready',
    epochs=100,
    patience=15,
    warmup_epochs=10,
    workers=4,

    imgsz=256,
    batch=128,

    dropout=0.3,
    label_smoothing=0.1,
    weight_decay=0.001,

    mixup=0.1,
    degrees=15.0,
    scale=0.5,
    fliplr=0.5,
    hsv_h=0.0,
    hsv_s=0.0,
    hsv_v=0.0,

    device=0,   
    project="food101_cls_yolo_s",
    name="yolo26_food101cls_run",
    verbose=True,
    plots=True,
)
```
